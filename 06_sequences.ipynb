Let's do some interesting stuff with lists. 

```python
def make_tea(ing):
    for i in ing: 
        print("Put", i, "in.")
        
        
ingredients = ['water', 'milk', 'cinnamon', 'tea leaves']
make_tea(ingredients)
```

```python
seq = [1, 2, 8, 4, 6, 8, 9, 15]
print(seq)
print(type(seq))
```

```python
def count(s, value):
    """Count how many times value occurs in s"""
    counter = 0 
    
    for i in s:  
        if i == value : 
            counter = counter + 1
            
    return counter 
```

```python
count(seq, 88)
```

```python
count(seq, 8)
```

### Other methods of element selection 

```python
digits = [1, 5, 4, 6, 7, 9]
```

```python
digits
```

```python
digits[2] # exact single index which return one value at a time

# digits[ start_index : end_index ]

digits[1:5]
```

```python
digits[1:4]      # stops just before index: 3
```

```python
len(digits)
```

```python
digits[-4:-1]
```

```python
digits[3:] # stops just before index:3 
```
