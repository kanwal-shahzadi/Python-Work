a.  Write a Python program that takes a list of numbers from the user and outputs the maximum, minimum, and average.


```python
numbers = input("Enter list of numbers by comma seprated")
numbers = numbers.split(',')
lst = []
for i in numbers:
    lst.append(int(i))
print("Maximum: ", max(lst))
print("Minimum: ", min(lst))
print("Average: ", sum(lst) / len(lst) )
```

b.    Using matplot, plot a bar chart for the following sales data
products = ['A', 'B', 'C', 'D']
sales = [150, 200, 300, 250]

```python
import matplotlib.pyplot as plt

products = ['A', 'B', 'C', 'D']
sales = [150, 200, 300, 250]

plt.bar(products, sales)
plt.xlabel("Products")
plt.ylabel("Sales")
plt.title("Product Sales")

plt.show()
```

##### Q 3: Introduction to Machine Learning (Theory)
 1. Name 2 types of ML with one example for each?

1. Supervised Learning
The model is trained on labeled data (data where the answer is already known). It learns to map inputs to specific outputs.

    - Example: Spam Detection. The system is fed thousands of emails labeled "spam" or "not spam" until it can accurately categorize new emails on its own.

2. Unsupervised Learning
The model works with unlabeled data. It finds hidden patterns or structures within the data without being told what to look for.

    - Example: Customer Segmentation. A business uses an algorithm to group customers into "clusters" based on similar buying habits (e.g., "bargain hunters" vs. "luxury shoppers").

##### 2. Explain what a dataset is, Why do we split it into training and testing sets?

**Answer:** A dataset is a collection of information used to teach and evaluate an AI model. We split it into a training set for learning patterns and a testing set to check performance on unseen data. This split ensures the model can actually predict new results rather than just memorizing its lessons.
